# Braghen stats comparison

```python
from MonteStats import roll_x_by_n_dice, Braghenise, sum_dice_rolls
import numpy as np
import matplotlib.pyplot as plt
```

## Generate stats

### testing

```python
def generate_intervals (dice, width = 0.2):
    minimum = dice
    maximum = dice*6
    intervals = np.arange(minimum - width/2, maximum + 1 + width, 1)

    return intervals
```

```python
generate_intervals(1)
```

```python
x_dice = 2
hist_width = 0.2
Base2d6 = roll_x_by_n_dice(x_dice, 500000)
BaseRolls = sum_dice_rolls(Base2d6)

plt.hist(BaseRolls, bins=generate_intervals(x_dice), width=hist_width)
```

Stats are working, so let's see Braghen

```python
x_dice = 2
hist_width = 0.2
Brag2d6 = Braghenise(Base2d6)
BaseBrag = sum_dice_rolls(Brag2d6)

_ =plt.hist(BaseBrag, bins=generate_intervals(x_dice), width=hist_width)
```

### base stats

```python
experiment_total = 1000000
```

```python
Regular2d6 = roll_x_by_n_dice(2, experiment_total)
```

```python
Regular3d6 = roll_x_by_n_dice(3, experiment_total)
```

```python
Brag2d6 = Braghenise(Regular2d6)
Brag3d6 = Braghenise(Regular3d6)
```

```python
Reg2d6Sum = sum_dice_rolls(Regular2d6)
Reg3d6Sum = sum_dice_rolls(Regular3d6)
Brag2d6Sum = sum_dice_rolls(Brag2d6)
Brag3d6Sum = sum_dice_rolls(Brag3d6)
```

## Visuals

```python
width = 0.4
plt.rcParams['figure.figsize'] = (12, 9)
plt.subplot (2, 2, 1)
_ = plt.hist (Reg2d6Sum, bins=generate_intervals(2), width=hist_width)
plt.title ('Regular 2d6')
plt.subplot (2, 2, 2)
_ = plt.hist (Reg3d6Sum, bins=generate_intervals(3), width=hist_width)
plt.title ('Regular 3d6')
plt.subplot (2, 2, 3)
_ = plt.hist (Brag2d6Sum, bins=generate_intervals(2), width=hist_width)
plt.title ('Braghen 2d6')
plt.subplot (2, 2, 4)
_ = plt.hist (Brag3d6Sum, bins=generate_intervals(3), width=hist_width)
plt.title ('Braghen 3d6')
plt.suptitle ('Montecarlo stats after 1 000 000 experiments')
```
